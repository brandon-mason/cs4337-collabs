## 4. **Optical Flow Visualization and Prediction**

### Purpose
Optical flow estimates motion between consecutive frames in a video by calculating pixel displacements. This is a powerful tool for understanding object movement and predicting future positions.

### Use Cases
- **Visualizing Optical Flow**: Optical flow can be represented as a color map, where colors indicate the direction and magnitude of movement. This helps in understanding how objects move in a scene.
- **Predicting the Next Frame**: By using optical flow to warp the current frame, we can predict how the next frame will look based on motion. This is useful in video compression, video interpolation, and predicting future movements.

### Real-World Application
- In autonomous driving, optical flow helps the vehicle’s vision system understand the movement of pedestrians or other vehicles in its surroundings, which is crucial for safe navigation.

````python
import cv2
import matplotlib.pyplot as plt

# Load an image from file
image = cv2.imread('image1.jpg')

# Convert BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
````

### Channel Swapping

````python
# Swap Red and Blue channels
swapped_image = image.copy()
swapped_image[:, :, [0, 2]] = swapped_image[:, :, [2, 0]]

# Display the swapped image
swapped_rgb = cv2.cvtColor(swapped_image, cv2.COLOR_BGR2RGB)
plt.imshow(swapped_rgb)
plt.title('Swapped Red and Blue Channels')
plt.axis('off')
plt.show()
````

### Applying Tangent Operations

````python
import numpy as np

# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply tangent operation
tangent_image = np.tan(gray_image.astype(float) / 255 * np.pi/4)
# division by 255 converts value to 0-1 value

# Normalize to 0-255
tangent_normalized = cv2.normalize(tangent_image, None, 0, 255, cv2.NORM_MINMAX)
tangent_uint8 = tangent_normalized.astype(np.uint8)

# Display the tangent image
plt.imshow(tangent_uint8, cmap='gray')
plt.title('Tangent Operation on Grayscale Image')
plt.axis('off')
plt.show()
````

#### What the np.tan?

When you apply `np.tan` (the tangent function) to an image in OpenCV, you're essentially applying the tangent of each pixel's intensity value. Let me break this down intuitively:

1. **Pixel Values in Images**: Each pixel in a grayscale image has an intensity value, usually ranging from 0 to 255, where 0 represents black, 255 represents white, and values in between represent shades of gray.

2. **Tangent Function (`np.tan`)**: The tangent function is periodic and grows rapidly as it approaches odd multiples of 90 degrees (or π/2 radians). It takes any input value and returns its tangent.

3. **Effect on Image**:
   - When `np.tan` is applied to each pixel, it computes the tangent of that pixel's intensity value. Since pixel values are usually integers (0–255), taking the tangent of these values transforms them, and the result can vary significantly.
   - The tangent function produces values that can grow infinitely large near certain input values, which may result in very large (or very small) pixel values. These values may get clipped when represented as image pixels.

4. **Result**:
   - In practical terms, applying `np.tan` to an image will result in an image with **extreme brightness shifts**, where pixel values can blow up around certain ranges, leading to very bright or very dark regions.
   - Because tangent behaves wildly around specific angles, the resulting image will often look highly distorted, with high-frequency changes between dark and bright regions.

5. **Clipping**: Since pixel values in images need to remain in a valid range (e.g., 0–255 for 8-bit images), the transformed values would be clipped to this range, which may lead to a loss of detail or exaggerated contrast in certain areas of the image.

In summary, applying `np.tan` to an image introduces nonlinear, rapid changes in pixel intensity, causing distortions and extreme contrasts. It's not a typical image transformation, but can be used to create special effects in some creative applications.

#### What the cv2.normalize?

`cv2.normalize` is a function in OpenCV that scales and shifts the pixel values of an image to a specific range, typically used to adjust the contrast or make pixel values consistent across images for better comparison or processing. It performs normalization of pixel intensities, which is crucial for tasks like feature scaling in machine learning or image enhancement.

Here’s how it works and its effect, explained intuitively:

##### Function Signature

```python
cv2.normalize(src, dst, alpha, beta, norm_type, dtype)
```

- **`src`**: Input image (the image that you want to normalize).
- **`dst`**: Output image (the result after normalization, which can overwrite the source image if set to `None`).
- **`alpha`**: The lower bound of the normalization range (e.g., 0).
- **`beta`**: The upper bound of the normalization range (e.g., 255).
- **`norm_type`**: The type of normalization to be applied. The most common types are:
  - `cv2.NORM_MINMAX`: Normalizes pixel values to the range `[alpha, beta]`.
  - `cv2.NORM_L2`, `cv2.NORM_INF`, and others (used for more specialized normalization).
- **`dtype`**: Data type of the output image, typically the same as the input image type.

##### How it Works:

1. **Min-Max Normalization (`cv2.NORM_MINMAX`)**:
   - **Idea**: It adjusts the pixel values such that the minimum pixel intensity in the image becomes `alpha` and the maximum pixel intensity becomes `beta`.
   - **Formula**: $\text{normalized_pixel} = \alpha + \frac{(pixel - min\_value)}{(max\_value- min\_value)} \times (\beta - \alpha)$

     Here, `min_value` and `max_value` are the minimum and maximum pixel intensities in the input image.
   - **Result**: If you normalize an image with a pixel intensity range from 50 to 200 to the range [0, 255], the minimum value (50) will become 0 and the maximum value (200) will become 255. All other pixel values will be scaled proportionally between 0 and 255.
   
   This normalization improves contrast by stretching the range of pixel intensities.

2. **L2 Normalization (`cv2.NORM_L2`)**:
   - **Idea**: Each pixel value is scaled such that the total sum of squared pixel values equals 1. This type of normalization is useful when you want to normalize an image's brightness while keeping the overall image intensity consistent.
   - **Formula**:
    $
    \text{normalized\_pixel} = \frac{\text{pixel}}{\sqrt{\sum (\text{all pixels}^2)}}
    $
   - **Result**: The output image will have pixel values that are proportionally scaled down, preserving the overall structure of the image but making its overall intensity consistent.

3. **L1 Normalization (`cv2.NORM_L1`)**:
   - **Idea**: The pixel values are scaled such that the sum of absolute pixel values equals 1.
   - **Formula**:
    $
    \text{normalized\_pixel} = \frac{\text{pixel}}{\sum (\text{all pixels})}
    $
   - **Result**: This is useful for scaling down pixel values while preserving the overall distribution of pixel intensities in the image.

##### Key Intuitive Effects:

- **Contrast Enhancement**: If your image has pixel values clustered within a narrow range (e.g., 100-150), normalizing it to a wider range like [0, 255] will stretch these values out, making dark areas darker and bright areas brighter, improving the contrast.
  
- **Consistency**: By normalizing multiple images to the same range, you ensure they have the same pixel intensity distribution, which is important for machine learning models or computer vision tasks where uniform input is needed.

- **Brightness Adjustments**: L2 or L1 normalization ensures that images with varying brightness levels are scaled to the same overall brightness, without altering the relative differences between pixel intensities.

## 2. Handling Multiple Images: Stacking and Permuting

### Objective
Learn to stack images horizontally, vertically, and in an nxm grid. Perform transpose operations to manipulate the arrangement of the images.

### Loading Multiple Images



````python
import os

# List of image filenames
image_files = ['image1.jpg', 'image2.jpg', 'image3.jpg', 'image4.jpg']

# # Load images into a list
images = [cv2.imread(file) for file in image_files]
````

````python
len(images)
````

````python
for img in images:
  print(img.shape, img.dtype)
````

````python
images = [cv2.resize(img, dsize = (512, 512)) for img in images]
````

````python
for img in images:
  print(img.shape, img.dtype)
````

### Stacking Images Horizontally and Vertically

````python
# Horizontal Stacking
horizontal_stack = cv2.hconcat(images)
horizontal_rgb = cv2.cvtColor(horizontal_stack, cv2.COLOR_BGR2RGB)
print(f"{horizontal_rgb.shape = }")
plt.imshow(horizontal_rgb)
plt.title('Horizontally Stacked Images')
plt.axis('off')
plt.show()
````

````python
# Vertical Stacking
vertical_stack = cv2.vconcat(images)
vertical_rgb = cv2.cvtColor(vertical_stack, cv2.COLOR_BGR2RGB)
plt.imshow(vertical_rgb)
plt.title('Vertically Stacked Images')
plt.axis('off')
plt.show()
````

### Creating a Grid Layout (nxm)

````python
def create_image_grid(images, rows, cols):
    img_height, img_width = images[0].shape[:2]
    resized_images = [cv2.resize(img, (img_width, img_height)) for img in images]

    grid_rows = [cv2.hconcat(resized_images[i*cols:(i+1)*cols]) for i in range(rows)]
    return cv2.vconcat(grid_rows)

rows, cols = 2, 2
grid_image = create_image_grid(images, rows, cols)
grid_rgb = cv2.cvtColor(grid_image, cv2.COLOR_BGR2RGB)
plt.imshow(grid_rgb)
plt.title('2x2 Image Grid')
plt.axis('off')
plt.show()
````

### Different strategies? Transpose and reshape?

````python
import numpy as np
image_batch = np.stack(images)
image_batch.shape
````

````python
# Step 1: Display the images in columns
batch_in_col = image_batch.reshape(-1, 512, 3)
print(f"{batch_in_col.shape = }")
plt.imshow(batch_in_col)
plt.axis('off')  # Turn off axis
plt.show()
````

````python
# Step 2: Display the images in rows
batch_transposed = image_batch.transpose(1,0,2,3)
print(f"{batch_transposed.shape = }")

batch_in_row = batch_transposed.reshape(512, -1, 3)
print(f"{batch_in_row.shape = }")

plt.imshow(batch_in_row)
plt.axis('off')  # Turn off axis
plt.show()
````

````python
# Display the images in grid
# Step 1: Reshape into a grid of 2x2 images
split_batch = image_batch.reshape(2, 2, 512, 512, 3)  # Shape (2, 2, 512, 512, 3)
print(f"{split_batch.shape = }")

# Step 2: Transpose to interleave rows and columns correctly
transposed_split = split_batch.transpose(0, 2, 1, 3, 4)  # Shape (2, 512, 2, 512, 3)
print(f"{transposed_split.shape = }")

B, H, D, W, C = transposed_split.shape

# Step 3: Reshape the array into a single large image for plotting (2 rows, 1024 width)
reshaped_row = transposed_split.reshape(B * H, D * W, 3)  # Shape (1024, 1024, 3)
print(f"{reshaped_row.shape = }")

# Step 4: Plot the reshaped grid
plt.imshow(reshaped_row)
plt.axis('off')  # Turn off axis
plt.show()
````

### Transpose Operations

````python
# Transpose (flip) each image horizontally
transposed_images = [cv2.transpose(img) for img in images]

# Horizontal stacking of transposed images
transposed_stack = cv2.hconcat(transposed_images)
transposed_rgb = cv2.cvtColor(transposed_stack, cv2.COLOR_BGR2RGB)
plt.imshow(transposed_rgb)
plt.title('Transposed (Flipped) Images')
plt.axis('off')
plt.show()
````
